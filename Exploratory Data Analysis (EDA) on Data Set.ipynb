The dataset has no duplicated rows. Here are 9 numerical features of type float64, 18 int type and the rest are object types (some of them have to be converted to exact type). We can see that numerical features have uncetainity. 

we have datasets in which different columns have different units â€“ like one column can be in kilograms, while another column can be in centimeters. Furthermore, we can have columns like Salary which can range from 85,000 to 1100,000, and even more; while an age column which can range from 0 to 100(at the most). Thus, Income is about 1,000 times larger than age.

But how can we be sure that the model treats both these variables equally? When we feed these features to the model as is, there is every chance that the Salary will influence the result more due to its larger value. But this doesnâ€™t necessarily mean it is more important as a predictor. So, to give importance to both Age, and Salary, we need feature transformation.
